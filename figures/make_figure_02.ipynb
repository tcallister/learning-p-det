```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.style.use("plotting.mplstyle")
import pandas as pd
from matplotlib.lines import Line2D

cbbh = '#1f78b4'
cnsbh = '#33a02c'
cbns = '#e31a1c'
cmisc = '#6a3d9a'
```

```python
train_data_bbh = pd.read_hdf('./../data/training_data/bbh_training_data.hdf')
certain_data_bbh = pd.read_hdf('./../data/training_data/rpo3-bbh-certain-formatted.hdf')
hopeless_data_bbh = pd.read_hdf('./../data/training_data/rpo3-bbh-hopeless-formatted.hdf')

train_data_bns = pd.read_hdf('./../data/training_data/bns_training_data.hdf')
certain_data_bns = pd.read_hdf('./../data/training_data/rpo3-bns-certain-formatted.hdf')
hopeless_data_bns = pd.read_hdf('./../data/training_data/rpo3-bns-hopeless-formatted.hdf')

train_data_nsbh = pd.read_hdf('./../data/training_data/nsbh_training_data.hdf')
certain_data_nsbh = pd.read_hdf('./../data/training_data/rpo3-nsbh-certain-formatted.hdf')
hopeless_data_nsbh = pd.read_hdf('./../data/training_data/rpo3-nsbh-hopeless-formatted.hdf')

hopeless_data_misc = pd.read_hdf('./../data/training_data/rpo3-combined-hopeless-formatted.hdf')
```

```python
for df in [train_data_bbh,certain_data_bbh,hopeless_data_bbh,
          train_data_bns,certain_data_bns,hopeless_data_bns,
          train_data_nsbh,certain_data_nsbh,hopeless_data_nsbh,
          hopeless_data_misc]:
    
    df['chirp_mass_source'] = df.chirp_mass_detector/(1.+df.redshift)
```

```python
fig = plt.figure(figsize=(10,3))

##########################
ax = fig.add_subplot(131)
ax.set_rasterization_zorder(1)
##########################

# Standard
ax.scatter(train_data_bbh.luminosity_distance,
           train_data_bbh.chirp_mass_source,
           s=0.1,alpha=0.05,c=cbbh)
ax.scatter(train_data_nsbh.luminosity_distance,
           train_data_nsbh.chirp_mass_source,
           s=0.1,alpha=0.05,c=cnsbh)
ax.scatter(train_data_bns.luminosity_distance,
           train_data_bns.chirp_mass_source,
           s=0.1,alpha=0.05,c=cbns)

# Hopeless
ax.scatter(hopeless_data_bbh.luminosity_distance,
           hopeless_data_bbh.chirp_mass_source,
           s=1,alpha=0.01,c='lightgrey',zorder=-1)
ax.scatter(hopeless_data_nsbh.luminosity_distance,
           hopeless_data_nsbh.chirp_mass_source,
           s=1,alpha=0.01,c='lightgrey',zorder=-1)
ax.scatter(hopeless_data_bns.luminosity_distance,
           hopeless_data_bns.chirp_mass_source,
           s=1,alpha=0.01,c='lightgrey',zorder=-1)
ax.scatter(hopeless_data_misc.luminosity_distance,
           hopeless_data_misc.chirp_mass_source,
           s=1,alpha=0.01,c='lightgrey',zorder=-1)

# Certain
ax.scatter(certain_data_bbh.luminosity_distance,
           certain_data_bbh.chirp_mass_source,
           s=1,alpha=0.01,c='lightgrey',zorder=-1)
ax.scatter(certain_data_nsbh.luminosity_distance,
           certain_data_nsbh.chirp_mass_source,
           s=1,alpha=0.01,c='lightgrey',zorder=-1)
ax.scatter(certain_data_bns.luminosity_distance,
           certain_data_bns.chirp_mass_source,
           s=1,alpha=0.01,c='lightgrey',zorder=-1)

ax.set_title('Pipeline Injections')
ax.set_xlabel('Luminosity Distance [Gpc]')
ax.set_ylabel('Source-Frame Chirp Mass [$M_\odot$]')
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlim(1e-2,30)
ax.set_ylim(6e-1,100)

bbh_circle = Line2D([0], [0], marker='o', color='w', label='BBH',
                        markerfacecolor=cbbh, markersize=5)
nsbh_circle = Line2D([0], [0], marker='o', color='w', label='NSBH',
                        markerfacecolor=cnsbh, markersize=5)
bns_circle = Line2D([0], [0], marker='o', color='w', label='BNS',
                        markerfacecolor=cbns, markersize=5)
plt.legend(handles=[bbh_circle,nsbh_circle,bns_circle],loc='upper left',handlelength=1)

##########################
ax = fig.add_subplot(132)
ax.set_rasterization_zorder(1)
########################## 

# Standard
ax.scatter(train_data_bbh.luminosity_distance,
           train_data_bbh.chirp_mass_source,
           s=0.1,alpha=0.01,c='lightgrey',zorder=-1)
ax.scatter(train_data_nsbh.luminosity_distance,
           train_data_nsbh.chirp_mass_source,
           s=0.1,alpha=0.01,c='lightgrey',zorder=-1)
ax.scatter(train_data_bns.luminosity_distance,
           train_data_bns.chirp_mass_source,
           s=0.1,alpha=0.01,c='lightgrey',zorder=-1)

# Hopeless
ax.scatter(hopeless_data_bbh.luminosity_distance,
           hopeless_data_bbh.chirp_mass_source,
           s=1,alpha=0.01,c=cbbh)
ax.scatter(hopeless_data_nsbh.luminosity_distance,
           hopeless_data_nsbh.chirp_mass_source,
           s=1,alpha=0.01,c=cnsbh)
ax.scatter(hopeless_data_bns.luminosity_distance,
           hopeless_data_bns.chirp_mass_source,
           s=1,alpha=0.01,c=cbns)
ax.scatter(hopeless_data_misc.luminosity_distance,
           hopeless_data_misc.chirp_mass_source,
           s=1,alpha=0.01,c=cmisc)

# Certain
ax.scatter(certain_data_bbh.luminosity_distance,
           certain_data_bbh.chirp_mass_source,
           s=1,alpha=0.01,c='lightgrey',zorder=-1)
ax.scatter(certain_data_nsbh.luminosity_distance,
           certain_data_nsbh.chirp_mass_source,
           s=1,alpha=0.01,c='lightgrey',zorder=-1)
ax.scatter(certain_data_bns.luminosity_distance,
           certain_data_bns.chirp_mass_source,
           s=1,alpha=0.01,c='lightgrey',zorder=-1)

ax.set_title("``Hopeless'' Injections")
ax.set_xlabel('Luminosity Distance [Gpc]')
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlim(1e-2,30)
ax.set_ylim(6e-1,100)
ax.set_yticklabels([])

bbh_circle = Line2D([0], [0], marker='o', color='w', label='BBH',
                        markerfacecolor=cbbh, markersize=5)
nsbh_circle = Line2D([0], [0], marker='o', color='w', label='NSBH',
                        markerfacecolor=cnsbh, markersize=5)
bns_circle = Line2D([0], [0], marker='o', color='w', label='BNS',
                        markerfacecolor=cbns, markersize=5)
misc_circle = Line2D([0], [0], marker='o', color='w', label='Misc.',
                        markerfacecolor=cmisc, markersize=5)
plt.legend(handles=[bbh_circle,nsbh_circle,bns_circle,misc_circle],loc='upper left',handlelength=1)

##########################
ax = fig.add_subplot(133)
ax.set_rasterization_zorder(1)
########################## 

# Standard
ax.scatter(train_data_bbh.luminosity_distance,
           train_data_bbh.chirp_mass_source,
           s=0.1,alpha=0.01,c='lightgrey',zorder=-1)
ax.scatter(train_data_nsbh.luminosity_distance,
           train_data_nsbh.chirp_mass_source,
           s=0.1,alpha=0.01,c='lightgrey',zorder=-1)
ax.scatter(train_data_bns.luminosity_distance,
           train_data_bns.chirp_mass_source,
           s=0.1,alpha=0.01,c='lightgrey',zorder=-1)

# Hopeless
ax.scatter(hopeless_data_bbh.luminosity_distance,
           hopeless_data_bbh.chirp_mass_source,
           s=1,alpha=0.01,c='lightgrey',zorder=-1)
ax.scatter(hopeless_data_nsbh.luminosity_distance,
           hopeless_data_nsbh.chirp_mass_source,
           s=1,alpha=0.01,c='lightgrey',zorder=-1)
ax.scatter(hopeless_data_bns.luminosity_distance,
           hopeless_data_bns.chirp_mass_source,
           s=1,alpha=0.01,c='lightgrey',zorder=-1)
ax.scatter(hopeless_data_misc.luminosity_distance,
           hopeless_data_misc.chirp_mass_source,
           s=1,alpha=0.01,c='lightgrey',zorder=-1)

# Certain
ax.scatter(certain_data_bbh.luminosity_distance,
           certain_data_bbh.chirp_mass_source,
           s=1,alpha=0.01,c=cbbh)
ax.scatter(certain_data_nsbh.luminosity_distance,
           certain_data_nsbh.chirp_mass_source,
           s=1,alpha=0.01,c=cnsbh)
ax.scatter(certain_data_bns.luminosity_distance,
           certain_data_bns.chirp_mass_source,
           s=1,alpha=0.01,c=cbns)

ax.set_title("``Certain'' Injections")
ax.set_xlabel('Luminosity Distance [Gpc]')
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlim(1e-2,30)
ax.set_ylim(6e-1,100)
ax.set_yticklabels([])

bbh_circle = Line2D([0], [0], marker='o', color='w', label='BBH',
                        markerfacecolor=cbbh, markersize=5)
nsbh_circle = Line2D([0], [0], marker='o', color='w', label='NSBH',
                        markerfacecolor=cnsbh, markersize=5)
bns_circle = Line2D([0], [0], marker='o', color='w', label='BNS',
                        markerfacecolor=cbns, markersize=5)
plt.legend(handles=[bbh_circle,nsbh_circle,bns_circle],loc='upper left',handlelength=1)

plt.tight_layout()
plt.savefig('figure_02.pdf',dpi=400,bbox_inches='tight')
plt.show()
```

```python
fig = plt.figure(figsize=(3,10))

s = 0.15
a = 0.1

#########################
# Col 1: Dist
#########################

ax = fig.add_subplot(411)
ax.set_rasterization_zorder(3)

for df in [train_data_bbh,certain_data_bbh,hopeless_data_bbh,
          train_data_bns,certain_data_bns,hopeless_data_bns,
          train_data_nsbh,certain_data_nsbh,hopeless_data_nsbh,
          hopeless_data_misc]:
    
    ax.scatter(df.luminosity_distance[df['detected']==1],df.chirp_mass_source[df['detected']==1],
               s=s,alpha=a,c=cbbh,zorder=2,marker='.')
    ax.scatter(df.luminosity_distance[df['detected']==0],df.chirp_mass_source[df['detected']==0],
               s=s,alpha=a,c=cbns,zorder=1,marker='.')
    
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlim(1e-2,30)
    ax.set_ylim(6e-1,100)
    ax.set_xticklabels([])
    ax.set_ylabel('$\mathcal{M}_c$',fontsize=12)
    
ax = fig.add_subplot(412)
ax.set_rasterization_zorder(2)

for df in [train_data_bbh,certain_data_bbh,hopeless_data_bbh,
          train_data_bns,certain_data_bns,hopeless_data_bns,
          train_data_nsbh,certain_data_nsbh,hopeless_data_nsbh,
          hopeless_data_misc]:
    
    ax.scatter(df.luminosity_distance[df['detected']==1],df.q[df['detected']==1],
               s=s,alpha=a,c=cbbh,zorder=2,marker='.')
    ax.scatter(df.luminosity_distance[df['detected']==0],df.q[df['detected']==0],
               s=s,alpha=a,c=cbns,zorder=1,marker='.')
    
    ax.set_xscale('log')
    ax.set_xlim(1e-2,30)
    ax.set_ylim(0,1)
    ax.set_xticklabels([])
    ax.set_ylabel('$q$',fontsize=12)
    
ax = fig.add_subplot(413)
ax.set_rasterization_zorder(2)

for df in [train_data_bbh,certain_data_bbh,hopeless_data_bbh,
          train_data_bns,certain_data_bns,hopeless_data_bns,
          train_data_nsbh,certain_data_nsbh,hopeless_data_nsbh,
          hopeless_data_misc]:
    
    ax.scatter(df.luminosity_distance[df['detected']==1],df.cos_inclination[df['detected']==1],
               s=s,alpha=a,c=cbbh,zorder=2,marker='.')
    ax.scatter(df.luminosity_distance[df['detected']==0],df.cos_inclination[df['detected']==0],
               s=s,alpha=a,c=cbns,zorder=1,marker='.')
    
    ax.set_xscale('log')
    ax.set_xlim(1e-2,30)
    ax.set_ylim(-1,1)
    ax.set_xticklabels([])
    ax.set_ylabel('$\cos\iota$',fontsize=12)
    
ax = fig.add_subplot(414)
ax.set_rasterization_zorder(2)

for df in [train_data_bbh,certain_data_bbh,hopeless_data_bbh,
          train_data_bns,certain_data_bns,hopeless_data_bns,
          train_data_nsbh,certain_data_nsbh,hopeless_data_nsbh,
          hopeless_data_misc]:
    
    ax.scatter(df.luminosity_distance[df['detected']==1],df.Xeff[df['detected']==1],
               s=s,alpha=a,c=cbbh,zorder=2,marker='.')
    ax.scatter(df.luminosity_distance[df['detected']==0],df.Xeff[df['detected']==0],
               s=s,alpha=a,c=cbns,zorder=1,marker='.')
    
    ax.set_xscale('log')
    ax.set_xlim(1e-2,30)
    ax.set_ylim(-1,1)
    ax.set_xlabel('$D_L$')
    ax.set_ylabel('$\chi_\mathrm{eff}$',fontsize=12)
    
plt.tight_layout()
plt.savefig('test.pdf',dpi=400)
```
